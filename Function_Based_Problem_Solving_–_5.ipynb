Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_post_engagement(likes_list):
    total_likes = 0
    #calculates the likes and post status
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")


# Example Usage
likes = [200, 300, 250, 300]
analyze_post_engagement(likes)
```

Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def rainfall_checker(rainfall_data, required_level):
    #Calculates average rainfall and checks adequacy.
    total = 0
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print(f"Average Rainfall: {average}")
    print(f"Rainfall Status: {status}")


# Example Usage
rain_data = [70, 80, 65, 75]
rainfall_checker(rain_data, 70)
```

Problem 4: Social Media – Duplicate Account Detection

```python
def detect_duplicate_usernames(usernames):
    #Detects duplicate usernames using set.
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"

    print(f"Duplicate Accounts Found: {result}")


# Example Usage
users = ["john", "alice", "john", "mike"]
detect_duplicate_usernames(users)
```

Problem 6: Agriculture – Premium Crop Price Filter

```python
def premium_crop_filter(prices):

    #Filters premium crop prices above ₹2000.

    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print(f"Premium Crops: {premium_crops}")


# Example Usage
crop_prices = [1500, 2500, 1800, 3200]
premium_crop_filter(crop_prices)
```

Problem 7: System Monitoring – Application Health Checker

```python
def application_health_checker(error_count):
    #Evaluates system health based on error count.
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")


# Example Usage
application_health_checker(7)
```

Problem 9: E-Learning – Student Attendance Eligibility System

```python
def attendance_checker(attendance_list):
    #Calculates attendance percentage and checks exam eligibility.
    total_classes = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == 1:  # 1 means present
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print(f"Attendance Percentage: {percentage}")
    print(f"Exam Eligibility: {eligibility}")


# Example Usage
attendance_data = [1,1,0,1,1]
attendance_checker(attendance_data)
```

Problem 10: Smart Electricity Bill Analyzer

```python
def electricity_bill_calculator(units):
    """
    Calculates electricity bill based on slab rates
    and classifies usage.
    """
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print(f"Units Consumed: {units}")
    print(f"Total Bill: ₹{bill}")
    print(f"Usage Category: {usage}")


# Example Usage
electricity_bill_calculator(250)
```
